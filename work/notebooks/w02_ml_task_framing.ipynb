## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# Lane: Refresh / Content Opportunity Scoring
# ML task type: This is a SCORING/RANKING problem (with a classification
# component underneath). I am not predicting one single label - I am producing
# a ranked list of pages, ordered by how likely/worth they are for review.
# Under the hood, the core piece is a binary classification task:
# will this page be a good refresh/review candidate or not.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# Success metric: Precision@50 (and Precision@K generally)
# Why: A reviewer only has capacity to check a limited number of pages.
# Precision@50 asks: of the top 50 ranked pages, how many are actually true
# positives? This matches the real decision (limited review capacity) far
# better than overall accuracy would.
```
